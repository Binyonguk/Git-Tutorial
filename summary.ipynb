# **미세먼지가 스포츠 관중수에 미치는 영향**

1. 연구의 목적
- 코로나 이전, 대한민국 사회,경제,정치적으로 가장 큰 이슈는 미세먼지
- 2018~2019년 미세먼지 니쁨 이후, 청명한 날의 연속이지만,
- 코로나19 이후에도 지속적으로 미세먼지 이슈가 없을 것인가?
- 특히, 주로야외에서 이루지는 스포츠는 미세먼지 영향을 많이 받는다고 하나,
- 실제 미세먼지 농도의 변화가 스포츠 관중수 변화에 영향이 있는지 분석 필요
 
 

2. 분석방법
- 회귀분석으로 상관관계를 분석
- 독립변수(미세먼지, 초미세먼지 농도), 종속변수(관중수)
- 분석툴은 파이썬 활용

3. 연구방법

 1) 독립변수 : 미세먼지, 초미세먼지 농도
 - 자료 : 공공데이터포털, 서울시열린데이터 광장
 - 자료수집 : 정보공개포털 API 크롤링진행
 - 기간 : 2015년 1월 ~ 2019년 12월

 2) 종속변수 : 프로농구 관중수
 - 자료 : 프로스포츠 정보광장 
 - 자료수집 : 정보공개 제공 Exel File
 - 기간 : 2015년 1월 ~ 2019년 12월

5. 결론
 - 최초, 미세먼지 농도가 높을 수록 이동량이 감소하여, 프로농구 관중수에 영향이 있을 것으로 가정했으나,
 - 회기분석 결과 미세먼지 농도 변화에 따른 프로농구 관중수 영향력이 없음
 - 향후, 다양한 스포츠(프로야구, 프로배구) 등에 적용하고 분석 예정
